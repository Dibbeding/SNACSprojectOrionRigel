```python
#plot node metrics evaluation
fig, ((ax2,ax3), (ax5, ax6), (ax8,ax9)) = plt.subplots(3, 2)

fig.set_figheight(18)
fig.set_figwidth(13)


X = ['Set 1','Set 2','Set 3','Set 4', 'Set 5']
  
X_axis = np.arange(len(X))
  


ax2.bar(X_axis - 0.2, twitch_rad_est, 0.4, label = 'Estimated')
ax2.bar(X_axis + 0.2, twitch_rad_ex, 0.4, label = 'Exact')
ax2.set_xlabel("Selected nodes", fontsize=15)
ax2.set_ylabel("metric value", fontsize=15)
ax2.set_title("Twitch - Radius", fontsize=18)
ax2.legend(loc = 'lower right')

ax3.bar(X_axis - 0.2, twitch_diam_est, 0.4, label = 'Estimated')
ax3.bar(X_axis + 0.2, twitch_diam_ex, 0.4, label = 'Exact')
ax3.set_xlabel("Selected nodes", fontsize=15)
ax3.set_ylabel("metric value", fontsize=15)
ax3.set_title("Twitch - Diameter", fontsize=18)
ax3.legend(loc = 'lower right')

ax5.bar(X_axis - 0.2, reddit_rad_est, 0.4, label = 'Estimated')
ax5.bar(X_axis + 0.2, reddit_rad_ex, 0.4, label = 'Exact')
ax5.set_xlabel("Selected nodes", fontsize=15)
ax5.set_ylabel("metric value", fontsize=15)
ax5.set_title("Reddit - Radius", fontsize=18)
ax5.legend(loc = 'lower right')

ax6.bar(X_axis - 0.2, reddit_diam_est, 0.4, label = 'Estimated')
ax6.bar(X_axis + 0.2, reddit_diam_ex, 0.4, label = 'Exact')
ax6.set_xlabel("Selected nodes", fontsize=15)
ax6.set_ylabel("metric value", fontsize=15)
ax6.set_title("Reddit - Diameter", fontsize=18)
ax6.legend(loc = 'lower right')

ax8.bar(X_axis - 0.2, deezer_rad_est, 0.4, label = 'Estimated')
ax8.bar(X_axis + 0.2, deezer_rad_ex, 0.4, label = 'Exact')
ax8.set_xlabel("Selected nodes", fontsize=15)
ax8.set_ylabel("metric value", fontsize=15)
ax8.set_title("Deezer - Radius", fontsize=18)
ax8.legend(loc = 'lower right')

ax9.bar(X_axis - 0.2, deezer_diam_est, 0.4, label = 'Estimated')
ax9.bar(X_axis + 0.2, deezer_diam_ex, 0.4, label = 'Exact')
ax9.set_xlabel("Selected nodes", fontsize=15)
ax9.set_ylabel("metric value", fontsize=15)
ax9.set_title("Deezer - Diameter", fontsize=18)
ax9.legend(loc = 'lower right')

matplotlib.rc('xtick', labelsize=13) 
matplotlib.rc('ytick', labelsize=13) 

plt.subplots_adjust(hspace = 0.3)

plt.show()
```

```python
#read error distribution data
with open('twitch_erdist.pickle', 'rb') as handle:
    twitch_erdist = pickle.load(handle)

with open('reddit_erdist.pickle', 'rb') as handle:
    reddit_erdist = pickle.load(handle)
    
with open('deezer_erdist.pickle', 'rb') as handle:
    deezer_erdist = pickle.load(handle)
    
with open('twitch_cdf.pickle', 'rb') as handle:
    twitch_dict = pickle.load(handle)
    twitch_cdf = twitch_dict['cdf']
    twitch_bins = twitch_dict['bins']
    
with open('reddit_cdf.pickle', 'rb') as handle:
    reddit_dict = pickle.load(handle)
    reddit_cdf = reddit_dict['cdf']
    reddit_bins = reddit_dict['bins']
    
with open('deezer_cdf.pickle', 'rb') as handle:
    deezer_dict = pickle.load(handle)
    deezer_cdf = deezer_dict['cdf']
    deezer_bins = deezer_dict['bins']
```

```python
#plot error distributions: CDFs
plt.plot(twitch_bins, twitch_cdf, label = 'Twitch')
plt.plot(reddit_bins, reddit_cdf, label = 'Reddit')
plt.plot(deezer_bins, deezer_cdf, label = 'Deezer')
plt.title('CDF of Relative Error', fontsize = 15)
plt.xlabel('Relative Error', fontsize = 12)
plt.ylabel('CDF', fontsize = 12)
plt.legend(fontsize = 12)
plt.xlim(0,0.75)
plt.show()
```

```python
# plot error distributions
plt.hist(reddit_erdist, bins = 100, alpha = 0.4, label = 'Reddit', color = 'red', 
                   histtype=u'step', linewidth=1.5)
plt.hist(twitch_erdist, bins = 100, alpha = 0.4, label = 'Twitch', color = 'blue', 
                   histtype=u'step', linewidth=1.5)
plt.hist(deezer_erdist, bins = 100, alpha = 0.4, label = 'Deezer', color = 'orange', 
                   histtype=u'step', linewidth=1.5)
plt.title('Relative Error distributions (on [0,1])', fontsize = 15)
plt.xlabel('RE', fontsize = 12)
plt.ylabel('#occurance', fontsize = 12)
plt.legend(fontsize = 12)
plt.show()
```

```python
# read centrality data
with open('reddit_tops_c.pickle', 'rb') as handle:
    reddit_tops_c = pickle.load(handle)

with open('reddit_tops_e.pickle', 'rb') as handle:
    reddit_tops_e = pickle.load(handle)
    
with open('deezer_tops_c.pickle', 'rb') as handle:
    deezer_tops_c = pickle.load(handle)

with open('deezer_tops_e.pickle', 'rb') as handle:
    deezer_tops_e = pickle.load(handle)
```

```python
#plot similarities between top nodes
plt.plot([10,20,30,40,50,60,70,80],reddit_tops_c, 'x-', label = 'closeness - Reddit')
plt.plot([10,20,30,40,50,60,70,80],reddit_tops_e, 'x-', label = 'eccentricity - Reddit')
plt.plot([10,20,30,40,50,60,70,80],deezer_tops_c, 'x-', label = 'closeness - Deezer')
plt.plot([10,20,30,40,50,60,70,80],deezer_tops_e, 'x-', label = 'eccentricity - Deezer')
plt.title('Similarity of top nodes', fontsize = 15)
plt.xlabel('top k', fontsize = 12)
plt.ylabel('similarity', fontsize = 12)
plt.legend()
plt.show()
```
